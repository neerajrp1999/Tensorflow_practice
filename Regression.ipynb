```python
import tensorflow as tf
import numpy as np
```

```python
my_xdata=tf.Variable(tf.random.uniform(shape=(50,),maxval=1000,dtype=tf.int32))
```

```python
my_ydata=my_xdata*13+40
my_xdata=np.asarray(my_xdata)
my_ydata=np.asarray(my_ydata)
```

```python
my_xdata , my_ydata.ndim
```

```python
#1
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation='relu'),
    tf.keras.layers.Dense(10,activation='relu'),
    tf.keras.layers.Dense(1,activation='relu')
    ]
)
model.compile(
    optimizer=tf.keras.optimizers.Adam(lr=0.0001),
    loss=tf.keras.losses.mae,
    metrics=['mae']
)
model.fit( my_xdata,my_ydata,epochs=1000)
```

```python
model.predict([50])
```
